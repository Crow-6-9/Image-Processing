```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('gg.jpeg' )
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# load the image and convert it to grayscale
gray=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
```

```python
#calclate the histogram
hist=cv2.calcHist([gray],[0],None,[256],[0,256])
# plt the histogram
plt.figure()
plt.title("Graysacle Histogram")
plt.xlabel("Bins")
plt.ylabel("No.of Pixels")
plt.plot(hist)
plt.xlim([0,256])
plt.show()
```

```python
#Equalized the histogram
eq_hist=cv2.equalizeHist(gray)
#calclate the histogram
hist=cv2.calcHist([eq_hist],[0],None,[256],[0,256])
# plt the histogram
plt.figure()
plt.title("Equalized Graysacle Histogram")
plt.xlabel("Bins")
plt.ylabel("No.of Pixels")
plt.plot(hist)
plt.xlim([0,256])
plt.show()
```

```python
fig,axs =plt.subplots(nrows=1, ncols=2, figsize=(5,5))
axs[0].imshow(gray)
axs[0].set_title('Gray Image')
axs[1].imshow(eq_hist)
axs[1].set_title('Equalized Gray Image')
```
