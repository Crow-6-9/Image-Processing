```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('gg.jpeg')
img2= cv2.imread('gg.jpeg')
x=plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
import numpy as np
 
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

sobel_x = cv2.Sobel(gray, cv2.CV_64F , 1 , 0, ksize=3)
sobel_y = cv2.Sobel(gray, cv2.CV_64F , 0 , 1, ksize=3)

magnitude = np.sqrt(sobel_x**2 + sobel_y**2)
angle = np.arctan2(sobel_y,sobel_x)

magnitude = (magnitude / np.max(magnitude)*255).astype(np.uint8)

_,binary_image = cv2.threshold(magnitude, 50,255, cv2.THRESH_BINARY)
```

```python
fig,axs = plt.subplots(nrows=1,ncols =2 , figsize=(5,5))

axs[0].imshow( cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[1].imshow(binary_image)
axs[1].set_title('Binary Image')
```

```python
blurred = cv2.GaussianBlur(cv2.cvtColor(img, cv2.COLOR_BGR2RGB), (5, 5), 0)

# Apply Laplacian edge detection
lap_edges = cv2.Laplacian(blurred, cv2.CV_64F)

# Normalize the Laplacian edges
lap_edges_normalized = cv2.normalize(lap_edges, None, 0, 255, cv2.NORM_MINMAX, cv2.CV_8U)
```

```python
fig, axs = plt.subplots(nrows=1 , ncols=2, figsize=(5,5))

axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title('Original Image')
axs[1].imshow(lap_edges_normalized)
axs[1].set_title('Laplacian Operator Image')
```
